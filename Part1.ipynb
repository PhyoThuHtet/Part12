```python
import bs4
import requests
import pandas as pd
import numpy as np
```

```python
# Calling the list of postal code toranto 
toronto = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
# Checking if the call is successful or not
toronto.status_code
```

```python
toronto_content = bs4.BeautifulSoup(toronto.content,'html.parser')
```

```python
table = toronto_content.find_all('table')[0]
```

```python
df = pd.read_html(str(table))[0]
```

```python
df
```

```python
df.columns= [x for x in ['Postcode','Borough','Neighbourhood']]
```

```python
df
```

```python
df = df.drop([0])
```

```python
replace = []
for i in df['Borough']:
    if (i == 'Not assigned'):
        replace.append(np.nan)
    else:
        replace.append(i)
df['Borough'] = replace
```

```python
df.dropna(axis = 0 ,inplace = True)
```
